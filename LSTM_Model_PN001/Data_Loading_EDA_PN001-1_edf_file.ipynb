```python
import mne
import pandas as pd

# Step 1: Load the EDF file
edf_path = 'data/PN00-1.edf'  # Replace with your actual path
raw = mne.io.read_raw_edf(edf_path, preload=True)

# Step 2: Get data and channel names
data, times = raw.get_data(return_times=True)
channel_names = raw.ch_names

# Step 3: Create a DataFrame (one row per time point)
df = pd.DataFrame(data.T, columns=channel_names)
df['time_sec'] = times  # Optional: include time in seconds

# Step 4: Save to CSV
df.to_csv('PN00-1.csv', index=False)

print("EDF file successfully converted to CSV.")
```

```python
import matplotlib.pyplot as plt
sample_channels = df.columns[:5]  # Pick first 5 EEG channels
df[sample_channels].plot(figsize=(15, 6), title='Sample EEG Signals')
plt.xlabel('Time Index')
plt.ylabel('Amplitude')
plt.show()
```

```python
df.plot(x='time_sec', y=sample_channels, figsize=(15, 6), title='EEG Signals Over Time')
plt.xlabel('Time (sec)')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(12, 6))
for ch in sample_channels:
    sns.kdeplot(df[ch], label=ch)
plt.title('Distribution of EEG Amplitudes')
plt.xlabel('Amplitude')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df[sample_channels].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Channels')
plt.show()
```

```python
#Power Spectral Density (PSD)
from scipy.signal import welch

fs = raw.info['sfreq']  # Sampling frequency

plt.figure(figsize=(15, 8))
for ch in sample_channels:
    f, Pxx = welch(df[ch], fs=fs, nperseg=1024)
    plt.semilogy(f, Pxx, label=ch)
plt.title('Power Spectral Density')
plt.xlabel('Frequency [Hz]')
plt.ylabel('PSD [V**2/Hz]')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
def bandpower(data, sf, band, window_sec=None):
    from scipy.signal import welch
    band = np.asarray(band)
    low, high = band
    if window_sec is not None:
        nperseg = int(window_sec * sf)
    else:
        nperseg = (2 / low) * sf
    freqs, psd = welch(data, sf, nperseg=int(nperseg))
    idx_band = np.logical_and(freqs >= low, freqs <= high)
    return np.trapz(psd[idx_band], freqs[idx_band])

sf = raw.info['sfreq']
bands = {'Delta': (0.5, 4), 'Theta': (4, 8), 'Alpha': (8, 13), 'Beta': (13, 30), 'Gamma': (30, 45)}

for ch in sample_channels:
    print(f'\nChannel: {ch}')
    for band in bands:
        power = bandpower(df[ch], sf, bands[band])
        print(f'{band} band power: {power:.2f}')
```

```python
# Check if any channel is flat (zero variance)
print(df[sample_channels].std())

# Plot boxplot for outlier detection
plt.figure(figsize=(12, 6))
df[sample_channels].plot(kind='box')
plt.title('Outlier Detection')
plt.show()
```
