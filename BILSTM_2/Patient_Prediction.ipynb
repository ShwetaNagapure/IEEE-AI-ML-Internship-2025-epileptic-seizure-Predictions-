```python
!pip install mne --quiet

import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.patches as patches
from tensorflow.keras.models import load_model
import mne
import warnings
warnings.filterwarnings("ignore")
from google.colab import drive
drive.mount('/content/drive')

# ----- Helper Functions -----
def load_edf(file_path):
    raw = mne.io.read_raw_edf(file_path, preload=True, verbose=False)
    raw.resample(256)
    return raw

def segment_eeg(raw, start, end, chunk_duration=5):
    sfreq = int(raw.info['sfreq'])
    data = raw.get_data()
    chunks = []
    for t in range(start, end - chunk_duration + 1, chunk_duration):
        s, e = t * sfreq, (t + chunk_duration) * sfreq
        chunk = data[:, s:e]
        norm = (chunk - np.mean(chunk, axis=1, keepdims=True)) / np.std(chunk, axis=1, keepdims=True)
        norm = np.nan_to_num(norm)
        chunks.append(norm.T)
    return chunks

def save_predictions_csv(file_name, preds_prob, preds_binary):
    df = pd.DataFrame({
        "chunk_index": np.arange(len(preds_binary)),
        "time_seconds": np.arange(len(preds_binary)) * 5,
        "seizure_probability": preds_prob,
        "seizure_predicted": preds_binary
    })
    csv_path = f'/content/drive/MyDrive/BILSTM_MODEL/predictions_{file_name}.csv'
    df.to_csv(csv_path, index=False)
    print(f"‚úÖ Saved predictions CSV: {csv_path}")

# ----- Load Model -----
model_path = '/content/drive/MyDrive/BILSTM_MODEL/final_seizure_model.h5'
model = load_model(model_path)
print("‚úÖ Model loaded successfully!")

# ----- File Paths and Patients -----
base_path = '/content/drive/MyDrive/BILSTM_MODEL/edf_files/'
pn00_files = [f'PN00-{i}.edf' for i in range(1, 6)]

# ----- Process Each File -----
for file_name in pn00_files:
    print(f"\nüîπ Processing {file_name}")
    file_path = os.path.join(base_path, file_name)

    try:
        raw = load_edf(file_path)
        total_duration = int(raw.n_times / raw.info['sfreq'])
        chunks = segment_eeg(raw, 0, total_duration)
        X = np.array(chunks)

        preds_prob = model.predict(X, verbose=0).flatten()
        preds_binary = (preds_prob > 0.5).astype(int)

        seizure_count = np.sum(preds_binary)
        print(f"Total Chunks: {len(preds_binary)} | Seizure Predicted Chunks: {seizure_count}")

        if seizure_count > 0:
            seizure_chunks = np.where(preds_binary == 1)[0]
            times_sec = seizure_chunks * 5
            print(f"‚ö†Ô∏è Seizure detected at chunk indices: {seizure_chunks}")
            print(f"‚è∞ Approximate seizure times (seconds): {times_sec}")
        else:
            print("‚úÖ No seizure detected in this recording.")

        # Plot seizure probability & prediction with shaded seizure chunks
        plt.figure(figsize=(12, 3))
        plt.plot(preds_prob, label='Seizure Probability', color='blue')
        plt.plot(preds_binary, label='Seizure Prediction (Threshold=0.5)', color='red', alpha=0.6)

        # Highlight seizure chunks with red shaded background
        for idx in np.where(preds_binary == 1)[0]:
            plt.gca().add_patch(
                patches.Rectangle((idx, 0), 1, 1.1, color='red', alpha=0.2)
            )

        plt.title(f"Seizure Prediction Timeline - {file_name}")
        plt.xlabel("Chunk Index (5 sec each)")
        plt.ylabel("Prediction / Probability")
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

        # Save predictions CSV for mentor review
        save_predictions_csv(file_name.replace('.edf', ''), preds_prob, preds_binary)

    except Exception as e:
        print(f"‚ùå Error processing {file_name}: {e}")

print("\n All done! Seizure prediction and detection completed for all files.")
```
